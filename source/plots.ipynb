```python
dirs = {'lstm_words': 8,
        'lstm_bpe': 10,
        'transformer_words': 8,
        'transformer_bpe': 19
       }
```

```python
train_perplexity = defaultdict(list)
val_perplrxity = defaultdict(list)
for d in dirs.keys():
    for epoch in range(dirs[d] + 1):
        json_path = '../models/' + d + '/metrics_epoch_{}.json'.format(epoch)
        with open(json_path) as f:
            j = json.load(f)
            train_perplexity[d].append(j['training_perplexity'])
            val_perplrxity[d].append(j['validation_perplexity'])
```

```python
train_df = pd.read_csv('../tb_metrics/run-transformer_bpe_log_train-tag-perplexity.csv')
val_df = pd.read_csv('../tb_metrics/run-transformer_bpe_log_validation-tag-perplexity.csv')
plt.figure(figsize=(10, 8))
plt.plot(train_df.Step, train_df.Value, label='Training perplexity')
plt.plot(val_df.Step, val_df.Value, label='Validation perplexity')
plt.xlabel("Epoch number")
plt.ylabel("Perplexity")
plt.title("Transformer with subwords")
plt.ylim(0, 50)
plt.legend()
plt.savefig('../img/transformer-bpe.png', dpi=300)
```
